## SELIC x IPCA

```python
DF_IPCA = data[77:]
DF_SELIC = data_selic[:424]
```

```python
df = pd.DataFrame({'Date': pd.to_datetime(DF_IPCA.index)}, index=None)
df['IPCA'] = DF_IPCA.values
df['SELIC'] = DF_SELIC.values
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month  
df
```

```python
import matplotlib.pyplot as plt

plt.plot(np.array(df['Year']),np.array(df['IPCA']),label="IPCA")
plt.plot(np.array(df['Year']),np.array(df['SELIC']),label="SELIC")


plt.title("IPCA x SELIC",fontsize=15)
plt.xlabel("Year",fontsize=13)
plt.ylabel("%",fontsize=13)
plt.legend()
plt.show()
```
